# Plotting Barplot using Seaborn

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# load dataset
tips_df = sns.load_dataset("tips")
print(tips_df.head())
```

```python
# Plot tips_df.day & tips_df.total_bill barplot
sns.barplot(x = tips_df.day, y = tips_df.total_bill)
plt.show()
```

```python
# Devide barplot using hue parameter
sns.barplot(x = tips_df.day, y = tips_df.total_bill, hue = tips_df.sex)
```

```python
# Pass dataset using data parameter
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df)
```

```python
# modify the order of day
order = ['Sun', 'Thur', 'Fri', 'Sat']

sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, order = order)
```

```python
#Modify hue order
hue_order = ['Female', 'Male']

sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, hue_order = hue_order)
```

```python
# estimate y variable value and then plot
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, estimator= np.max)
```

```python
# set barplot as horizontal
sns.barplot(y = 'day', x = 'total_bill', hue = 'sex', data = tips_df)
```

```python
# set color
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, color="g")
```

# sns.barplot() kwargs parameter
Seaborn bar plot Keyword arguments (kwargs) help to give an artistic look to the bar graph. You can pass plt.bar() keyword arguments to it. Check out the matplotlib bar char article.

Here, we are using some barplot kwargs like:

- alpha
- linestyle
- linewidth
- edgecolor
- etc

```python
# Keyword Arguments parameter
kwargs = {'alpha':0.9, 'linestyle':':', 'linewidth':5, 'edgecolor':'k'}

sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df,**kwargs)
```

```python
# Pass Keyword  argument as parameter
sns.set() # for darkgrid background
sns.barplot(x = 'day', y = 'total_bill', data = tips_df, alpha =.9, linestyle = "-.", linewidth = 3)
```

```python
# Axes parameter
sns.set()
ax = sns.barplot(x = 'day', y = 'total_bill',
           data = tips_df, alpha =.9, linestyle = "-.", linewidth = 3,
           palette="Set3")

ax.set(title = "Barplot of Tips DataFrame",
      xlabel = "Days",
      ylabel = "Total Bill")
```

# Full example

```python
# Example of Seaborn Barplot
sns.set()
plt.figure(figsize = (16,9))

sns.barplot(x = 'day', y = 'total_bill',
           data = tips_df, alpha =1, linestyle = "-.", linewidth = 3,
           palette="Set3")

plt.title("Barplot of Days and Total Bill", fontsize = 20)
plt.xlabel("Days", fontsize = 15)
plt.ylabel("Total Bill", fontsize = 15)

plt.savefig("Barplot of Days and Total Bill")
plt.show()
```

```python
# prompt: color bars with differentes colors

import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips_df = sns.load_dataset("tips")

# Create the barplot with different colors for each day
sns.barplot(x='day', y='total_bill', data=tips_df, palette="Set3")  # Use a categorical palette

plt.title("Barplot of Days and Total Bill with Different Colors")
plt.xlabel("Days")
plt.ylabel("Total Bill")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='day', y='total_bill', data=tips_df, estimator='mean', hue='sex', palette='Set1')

# Add title and labels
plt.title('Average Total Bill per Day by Gender')
plt.xlabel('Day of the Week')
plt.ylabel('Average Total Bill ($)')

# Show the plot
plt.show()
```
