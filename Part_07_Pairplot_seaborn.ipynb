```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
df = sns.load_dataset('iris')

# Create pairplot
sns.pairplot(df)
plt.show()
```

```python
# Hue (to color by category):
sns.pairplot(df, hue='species')
```

```python
# Kind (use KDE plots or regression lines):
sns.pairplot(df, kind='kde')  # or kind='scatter
```

```python
sns.pairplot(df, hue='species', markers=["o", "s", "D"], palette="husl")
```

```python
# pairpot with using x_var and y_var
sns.pairplot(df, x_vars=['sepal_length', 'sepal_width'], y_vars=['petal_length', 'petal_width'], hue='species')
```
